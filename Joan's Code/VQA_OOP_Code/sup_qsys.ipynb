```python
import pennylane as qml 
from pennylane import numpy as np
import matplotlib.pyplot as plt
```

```python
qubits = 1
s = 1
dev = qml.device("default.qubit", wires=qubits, shots = s)

@qml.qnode(dev)

def superposition():
    qml.Hadamard(wires = 0)
    return qml.probs()

superposition()
```

```python
y = superposition()
bit_strings = [f"{x:0{qubits}b}" for x in range(len(y))]
plt.bar(bit_strings, y, color = "#212121")
plt.xlabel("State label")
plt.ylabel("Probability Amplitude")
plt.title("bit probability amplitudes")
```

```python
qubits = 3
s = 10
dev = qml.device("default.qubit", wires=qubits, shots = s)

@qml.qnode(dev)
def superposition(qubits):
    for i in range(qubits):
        qml.Hadamard(wires = i)
    return qml.probs()

draw2 = qml.draw(superposition)   
print(draw2(qubits))
```

```python
y = superposition(qubits)
bit_strings = [f"{x:0{qubits}b}" for x in range(len(y))]
plt.bar(bit_strings, y, color = "#212121")
plt.xlabel("State label")
plt.ylabel("Probability Amplitude")
plt.title("bit probability amplitudes")
```
