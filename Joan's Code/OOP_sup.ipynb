```python
class Sup:
    
    def __init__(self, qubits, shots):
        self.qubits = qubits
        self.shots = shots
    
    def run(self):
        #wrapper
        dev = qml.device("default.qubit", wires=self.qubits, shots = self.shots)
        @qml.qnode(dev)
        #circuit
        def H(q):
            for i in range(self.qubits):
                qml.Hadamard(i)
            return [qml.expval(qml.PauliZ(i)) for i in range(self.qubits)]
            #return qml.sample()
        return (H(self.qubits))
```
